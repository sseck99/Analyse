```python
# Exercice_1 
import numpy as np 
import matplotlib.pyplot as plt 

note = np.random.randint(21, size= 50 )
print("la note de tout les etudiant est",note)

# calcule de le moyenne et le mediane et l'ecart-type

# POur la moyenne 
S = sum(note)
D = len(note) 
M = S/D
print("la moyenne de la classe est:",M)

# Pour la Mediane 
Me = np.median(note)
print("la mediane de la classe est:",Me)
# Pour l'ecart-type
ecart_type = np.std(note)
print("l'ecart-type de la classe est:",ecart_type)
# Le nombre d'etudiant ayant une note superieur ou egal a 10 

c = 0 
for i in note:
    if i >= 10:
        c+=1
print("le nombre etudiant qui on la moyenne est de:",c)

# Histogramme des notes 

plt.hist(note , bins=[0,2.5,5,7.5,10,12.5,15,17.5,20],edgecolor="black")
plt.axvline(Me , color="r", label= 'la mediane')
plt.axvline(M, color='y', label="la moyenne")
plt.xlabel("Note")
plt.ylabel("Densite")
plt.grid(axis='y', alpha=0.5)
plt.title("Histogramme des notes ")
plt.legend()
plt.show()
# Diagramme circulaire 
nombre_etudint= np.array([c,50-c])
tab= ['Etudiants reussi',
      'Etudiants echoue']
print
plt.title('la proportion d’étudiants ayant réussi  et échoué ')
plt.pie(nombre_etudint, labels=tab, autopct='%d%%')
plt.show()
```
